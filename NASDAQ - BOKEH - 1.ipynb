```python
sec=pd.read_csv("D:/DATASETS/NYSE/securities.csv")
tick=sec['Ticker symbol']
ticker=tick.unique()

del_tick=['HCN','PX','CBG','SPLS','ARNC','PCLN','LUK','LLL','SYMC','CELG','HRS','FOXA','FOX','LVLT','WFM','IR','DPS','BF.B','DOW',
         'TSO','BCR','SE','KORS','AET','CSRA','DLPH','VIAB','BBT','COH','APC','MJN','ESRX','RAI','RHT','CA','HCP','BRK.B','GGP',
          'HAR','YHOO','WYN','EVHC','TSS','TMK','BHI','LLTC','XL','ABT','MMM']

tickers=[x for x in ticker if x not in del_tick]
```

```python
stock_data=pd.DataFrame()
app_data=[]
for t in tickers:
    dat=yf.download(t,start="2019-01-01", end="2021-06-07",interval = "3mo")
    dat['TICKERS']=t
    print (t)
    app_data.append(dat)

stock_data=pd.concat(app_data) 
```

```python
stock_data.to_csv("D:/DATASETS/NYSE/NYSE_prices.csv")
st_df=pd.read_csv("D:/DATASETS/NYSE/NYSE_prices.csv")
st_df.head()
```

```python
st_df['Date']=pd.to_datetime(st_df['Date'])
st_df['Change']=st_df['Close']-st_df['Open']
st_df.head()
```

```python
ind_sec=sec[['Ticker symbol','GICS Sector','GICS Sub Industry','Security']]
all_df=pd.merge(st_df,ind_sec,how='inner',left_on='TICKERS', right_on='Ticker symbol')
all_df.head()
```

```python
all_df=all_df.dropna()
all_df.shape
```

```python
def fig(f):
    
    #Create the figure object
    #f = figure(x_axis_type='datetime')
    colors =['yellow', 'blue', 'green', 'brown', 'darkcyan', 'pink', 'orange','purple', 'grey', 'red']   
    #Style the plot area
    f.plot_width = 1200
    f.plot_height = 600
    f.background_fill_color = "black"
    #f.background_fill_alpha = 0.3

    #Style the title
    f.title.text_color = "black"
    f.title.text_font = "calibri"
    f.title.text_font_size = "20px"
    f.title.align = "center"

    #Style the axes
    f.xaxis.minor_tick_line_color = "black"
    f.yaxis.major_label_orientation = "vertical"
    f.xaxis.visible = True
    f.xaxis.minor_tick_in = -6
    f.xaxis.axis_label = "Time"
    f.yaxis.axis_label = "Price"
    f.axis.axis_label_text_color = "black"
    f.axis.major_label_text_color = "black"
    
    return colors,f
```

```python
def graph(df,f):
    
    symbols = df["TICKERS"].unique().tolist()
    TickMax={}
    for u in symbols:
        values = df[(df["TICKERS"] == u)]["Close"].tolist()
        #takes all the close values of one ticker and adds to an array - values
        TickMax[u] = max(values)
        #each tickers max value is taken for years 1995-2000

        top_10= list(dict(sorted(TickMax.items(),key=lambda v:v[1], #sort by key -> v[0] | sort by value->v[1]
                   reverse=True)[:10]).keys())
        #Dictionary with ticker and close values. Take only top 10 tickers.

        top = [i in top_10 for i in df["TICKERS"].tolist()]
        #get the values of those tickers and add that to the list
        df_all= df[top]

        symbols=df_all["TICKERS"].unique().tolist()
    colors,f=fig(f)
    for s,c,n in zip(symbols[:10],colors,symbols):
        dates=df_all[(df_all['TICKERS']==s)]['Date']
        values=df_all[(df_all['TICKERS']==s)]['Close']
        f.line(dates,values, color=c,line_width=4, legend_label=n)

    f.legend.location = "top_left" 

    show(f)
```

```python
df_2015=all_df[all_df['Date'].dt.year>2014]
df_2015.shape

#Define the output file path
output_file("2015-2020_Analysis.html")
f = figure(x_axis_type='datetime')
f.title.text = "2015-2020 Year Stock Analysis"
graph(df_2015,f)
```

```python
#Grouping the data by sector and date, to get mean of each sector for every month
sector_df=all_df.groupby(['Date','GICS Sector'])['Close'].mean().reset_index()
sect=sector_df['GICS Sector'].unique().tolist()

#Define the output file path
output_file("Sector_2000-2020_Analysis.html")

#Create the figure object
f1 = figure(x_axis_type='datetime')
colors,f1=fig(f1)
for s,c,n in zip(sect,colors,sect):
    dates=sector_df[(sector_df['GICS Sector']==s)]['Date']
    values=sector_df[(sector_df['GICS Sector']==s)]['Close']
    f1.line(dates,values,color=c,line_width=4, legend_label=n) 

f1.legend.location = "top_left" 
show(f1)
```
